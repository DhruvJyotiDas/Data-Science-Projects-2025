```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
X_train_flattened = X_train.reshape(len(X_train), 28*28)
X_test_flattened = X_test.reshape(len(X_test), 28*28)
```

```python
X_train_flattened.shape
X_test_flattened.shape
```

```python
# Its converted to a 2D array, where each row is an image and each column is a pixel value.
# For example, the first image is now represented as a 1D array of 784 pixel values.
X_train_flattened[0]
```

Training a Simple 2 layer Neural Network 

```python
model = keras.Sequential([keras.layers.Dense(10, input_shape=(784,), activation='softmax')])

model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)


model.fit(X_train_flattened, Y_train, epochs=5)
```
